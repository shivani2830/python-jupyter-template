### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)  # x values from 0 to 10 with more points for smoothness
y = x**2  # y values as the square of x

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='', linestyle='-', color='b', linewidth=2, label=r'$y = x^2$')

# Customize the plot
plt.title('Square Function Plot', fontsize=14, fontweight='bold')
plt.xlabel('X values', fontsize=12)
plt.ylabel('Y values (X squared)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)

# Show the plot
plt.show()
```
